```python
from skimage.io import imread, imshow
from skimage.feature import hog
from skimage.color import rgb2gray
from skimage.draw import polygon, line

from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.utils import shuffle
from sklearn.metrics import roc_auc_score, roc_curve

import numpy as np
from matplotlib import pyplot as plt
import glob
from pprint import pprint
from collections import defaultdict
from queue import LifoQueue

%matplotlib inline
```

## Часть 1

```python
train_faces = []
train_non_faces = []

for filename in glob.glob('./train/faces/*.png'):
    train_faces.append(imread(filename))
for filename in glob.glob('./train/non-faces/*.png'):
    train_non_faces.append(imread(filename))
```

```python
def extract_hog_features(imgs, pixels_per_cell=(6, 5)):
    hog_features = []
    for img in imgs:
        hog_features.append(hog(rgb2gray(img), orientations=9, cells_per_block=(1, 1), 
                                pixels_per_cell=pixels_per_cell))
    return hog_features
```

```python
def build_classifier(X, y, show_auc=False):
    alg = SVC(probability=True)
    cv_score = np.mean(cross_val_score(alg, X, y, n_jobs=-1, cv=5, scoring='roc_auc'))
    if show_auc:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
        alg.fit(X_train, y_train)
        y_pred = alg.predict_proba(X_test)
        y_test = np.array(y_test)
        fpr, tpr, thresholds = roc_curve(y_test, y_pred[:, 1])
        plt.plot(fpr, tpr)
        plt.show()
    alg.fit(X, y)
    return alg, cv_score
```

### Подбор параметра

```python
def search_params(faces, non_faces, params):
    scores = {}
    for pixels_per_cell in params:
        faces_features = extract_hog_features(faces, pixels_per_cell)
        non_faces_features = extract_hog_features(non_faces, pixels_per_cell)
        X = faces_features + non_faces_features
        y = [1 for _ in range(len(faces_features))] + [0 for _ in range(len(non_faces_features))]

        X, y = shuffle(X, y)
        _, score = build_classifier(X, y)
        scores[pixels_per_cell] = score
    pprint(scores)
    return max(scores, key=scores.get)
```

```python
%%time
faces_features = extract_hog_features(train_faces, pixels_per_cell=(6, 5))
non_faces_features = extract_hog_features(train_non_faces, pixels_per_cell=(6, 5))
X = faces_features + non_faces_features
y = [1 for _ in range(len(faces_features))] + [0 for _ in range(len(non_faces_features))]

X, y = shuffle(X, y)
clf, score = build_classifier(X, y, show_auc=True)
```

```python
%%time
best = search_params(train_faces, train_non_faces, params=[(6, 6), (6, 7), (7, 6), (8, 8), (5, 6), (6, 5), (5, 5)])
```

### Алгоритм поиска окном

Идем окном по картинке с шагом 2 по горизонтали и вертикали, по порогу оставляем точки с достаточно большим откликом. Если окна достаточно сильно пересекаются, то выделяем среди них связные компоненты, чтобы подавить немаксимумы.

```python
test_imgs = [imread(filename) for filename in glob.glob('./test/*.jpg')]
```

```python
def get_area(rectangle):
    return (rectangle[1] - rectangle[0]) * (rectangle[3] - rectangle[2])

def overlap_check(first, second, threshold):
    ovr = max(0, min(first[1], second[1]) - max(first[0], second[0])) * max(0, min(first[3], second[3]) - max(first[2], second[2]))
    if (float(ovr) / (get_area(first) + get_area(second)) > threshold):
        return True
    return False

def find_not_visited(marks):
    for vertex, mark in marks.items():
        if mark == 0:
            return vertex
    return None

def get_connected_components(graph):
    vertices = graph.keys()
    marks = {vertex: 0 for vertex in vertices}
    components = []
    
    while find_not_visited(marks) is not None:
        start_vertex = find_not_visited(marks)
        q = [start_vertex]

        cur_component = set()
        while q:
            cur_vertex = q.pop()
            marks[cur_vertex] = 1
            for adj_vertex in graph[cur_vertex]:
                cur_component.add(adj_vertex)
                if marks[adj_vertex] == 0:
                    q.append(adj_vertex)
        components.append(cur_component)
    return components
        
def sliding_window_search(img, clf, window_size, x_step, y_step, threshold=0.98):
    keypoints = {}
    img_x, img_y = img.shape[:-1]
    window_x, window_y = window_size
    for x in range(0, img_x, x_step):
        for y in range(0, img_y, y_step):
            if x + window_x < img_x and y + window_y < img_y:
                part = img[x: x + window_x, y: y + window_y]
                features = np.array(extract_hog_features([part]))
                confidence = clf.predict_proba(features)[:, 1]
                if confidence > threshold:
                    keypoints[(x, y)] = confidence
    
    graph = defaultdict(list)
    for first_point, _ in keypoints.items():
        for second_point, _ in keypoints.items():
            if second_point != first_point and overlap_check(
            [first_point[0], first_point[0] + window_x, first_point[1], first_point[1] + window_y],
            [second_point[0], second_point[0] + window_x, second_point[1], second_point[1] + window_y],
            threshold=0.3):
                graph[first_point].append(second_point)
                graph[second_point].append(first_point)
    connected_components = get_connected_components(graph)
    max_keypoints = {}
    for component in connected_components:
        confidences = {}
        for point in component:
            confidences[point] = keypoints[point]
        max_point = max(confidences, key=confidences.get)
        max_keypoints[max_point] = keypoints[max_point] 
    
    return max_keypoints
```

```python
def draw_window(img, points, window_shape):    
    copy_img = img.copy()
    for point in points:
        vertexes = [point, 
                    (point[0], point[1] + window_shape[1]), 
                    (point[0] + window_shape[0], point[1] + window_shape[1]),
                   (point[0] + window_shape[0], point[1])
                   ]
        for i in range(len(vertexes)):
            x_1, y_1 = vertexes[i]
            x_2, y_2 = vertexes[(i + 1) % 4]
            rr, cc = line(x_1, y_1, x_2, y_2)
            copy_img[rr, cc] = 255, 0, 0
    imshow(copy_img)
    plt.show()
```

```python
for img in test_imgs:
    points = sliding_window_search(img, clf, (32, 24), 2, 2)
    print(points.keys())
    draw_window(img, points.keys(), (32, 24))
```
